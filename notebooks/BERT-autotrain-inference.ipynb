```python
from transformers import AutoTokenizer, AutoModelForQuestionAnswering
from transformers import pipeline
import json

model_path = "./autotrain-bert-ex-qa1"

tokenizer = AutoTokenizer.from_pretrained(model_path)
model = AutoModelForQuestionAnswering.from_pretrained(model_path)
```

```python
qa = pipeline(
    "question-answering",
    model=model,
    tokenizer=tokenizer,
    device="mps"  # or -1 for CPU
)
```

```python
with open('../QA-dataset/autotrain_data/validation.json','r') as file:
    val=json.load(file)
```

```python
sample=val[90]
```

```python
test={'question': sample['question'],'context': sample['context']}
```

```python
qa(test)
```
